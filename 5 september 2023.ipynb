```python
def get_next_words_dict(text):
    # Convert the text to lowercase
    text = text.lower()
    
    # Split the text into words
    words = text.split()
    
    # Initialize an empty dictionary to store the next words for each unique word
    next_words_dict = {}
    
    # Iterate through the words to create the dictionary
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]
        
        # If the current_word is not in the dictionary, initialize an empty list for it
        if current_word not in next_words_dict:
            next_words_dict[current_word] = []
        
        # Append the next_word to the list for the current_word
        next_words_dict[current_word].append(next_word)
    
    # Sort the lists of next words in ascending order
    for key in next_words_dict:
        next_words_dict[key].sort()
    
    # Handle the last word in the text with an empty string as its next word
    last_word = words[-1]
    if last_word not in next_words_dict:
        next_words_dict[last_word] = ['']
    
    return next_words_dict

# Helper function to print the dictionary in alphabetical key order
def print_dict_in_key_order(a_dict):
    for key in sorted(a_dict.keys()):
        print(f"{key} : {a_dict[key]}")

# Example usage
text = 'Easy come easy go go easy go easy'
next_words_dict = get_next_words_dict(text)
print_dict_in_key_order(next_words_dict)
```
